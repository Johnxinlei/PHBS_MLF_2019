```python
from scipy.io import arff
import pandas as pd
```

```python
data = arff.loadarff('4year.arff')
```

```python
df = pd.DataFrame(data[0])
```

```python
df['bankruptcy'] = (df['class']==b'1')
```

```python
df
```

```python
sum(df.bankruptcy == True)
```

```python
df_new = df[['Attr1','Attr2','Attr7','Attr10','bankruptcy']]
new_col = ['X1', 'X2', 'X7','X10','class']
df_new.columns = new_col
```

```python
df_new
```

```python
df_new = df_new.fillna(df_new.mean())
```

```python
df_new
```

```python
df_new.isnull().any()
```

```python
print("mean of all companies:\n", df_new.loc[:,'X1':'X10'].mean())
print("std of all companies:\n", df_new.loc[:,'X1':'X10'].std())
```

```python
print("mean of bancrupt companies:\n",df_new[df_new['class'] == 1].loc[:,'X1':'X10'].mean())
print("std of bancrupt companies:\n",df_new[df_new['class'] == 1].loc[:,'X1':'X10'].std())
```

```python
print("mean of still-operating companies:\n",df_new[df_new['class'] == 0].loc[:,'X1':'X10'].mean())
print("std of still-operating companies:\n",df_new[df_new['class'] == 0].loc[:,'X1':'X10'].std())
```

```python
df_select = df_new[(df_new['X1'] < df_new.X1.mean()-df_new.X1.std()) & (df_new['X10'] < df_new.X10.mean()-df_new.X10.std())]
df_select
```

```python
len(df_select)
```

```python
ratio_bcp = sum(df_select['class']==1)/len(df_select)
ratio_bcp
```
